# Consecutive Coin Flips
#### or 
# "Why do we never deliver everything we plan?"

**Question:**  I'm going to throw a fair dice 1000 times while keeping a running sum: when I get heads I add one, tails I subtract one.  How many times should I expect to be at zero?

This question is from Don Reinertsen's _Principles of Product Development Flow_, where the author suggests it's the type of problem even trained mathematicians will have poor intution on.  Most people feel it's maybe 500 or 250\*. The solution is not trivial, and the answer is approximately **27**. 

This problem illustrates a random _process_ and the poor intuition comes from the human tendendency to believe in fairness over time, even when it doesn't exist.

Imagine that the first 5 flips were heads, and we are +5. From here we might believe there's _should_ be more tails coming, but of course, _randomness does not work that way_, the next flip has equal chance being a head or a tail.  But we do need to get more tails now to get back to zero, which is now less likely.

In fact, in this problem there is only a 50% chance you'll ever find yourself back at zero after the 500th flip (halfway).

Why is this interesting to product development?

Because our estimates are also subject to random variation. And because often we fnd ourselves behind schedule thinking we'll get back.  But the truth is that we "...cannot rely on randomness to correct the problems randomness creates."

```python
import random
```

```python
def rolls(n):
    """Make a random series of n rolls, +1 for heads, -1 for tails."""
    return [ random.choice([1,-1])for _ in range(n) ]
```

```python
three_sequences = pandas.DataFrame(data=[rolls(1000), rolls(1000), rolls(1000), [0 for _ in range(1000)]])
three_sequences.T.cumsum().plot(figsize=(12,6), legend=False)
```

Above are three iterations of 1000 tosses. We can see the large variances away from the zero line. We can also see how easy it is to get awat from zero and never come back.
